```python
### 7.a - Select and print 5 lines from the beginning of a price 
###     maximal column and calculate sum of the values, minimal value, 
###     value, median

mySecondList = cars.head(5)

print( mySecondList )

print("maximal price = ",mySecondList["price"].max())
print("minimal price = ",mySecondList["price"].min())

#print("median = ", mySecondList["price"].median())
#print("sum = ", mySecondList["price"].sum())
```

```python
sum = 0
maList = []

for i in range(0,5,1):
    ### warning : value is not a float data. It's a string data with a special caracter
    value = mySecondList.iloc[i,5]
       
    ### i clean the data
    value = value.replace("€","",1)  
    
    """ 
        To be honest, i don't understand why i can't make this command :  
                          value = float(value)
                          
        if i do that, i will have this error on my terminal : 
                  "could not convert string to float"                 
    """
    ### i slipt my data in order to delete the ","
    value = value.split(",")

    ### now, i recreat a float number without the original ","
    value = float(value[0]) + int(value[1])*0.01
    
    ### i save my new data on a list
    maList.append(value)
    
    ### i make a sum 
    sum = sum + value

### i add a new column with the clean data
mySecondList["PriceV2"] = maList

print(" => My Method : sum = ",sum," € ")     
print(" => the classic Method : sum = ",mySecondList["PriceV2"].sum()," € ")
print(" => the classic Methos : median = ", mySecondList["PriceV2"].median())
```
